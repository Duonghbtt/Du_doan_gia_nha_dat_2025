```python
# c·ªôt gi√° ti·ªÅn
import re
import numpy as np
import pandas as pd

def convert_price_to_billion(value):
    if pd.isna(value):
        return np.nan

    # L√†m s·∫°ch k√Ω t·ª± l·∫°
    text = str(value).lower().strip()
    text = re.sub(r"[^0-9a-zA-Z√Ä-·ªπ,.\s]", "", text)
    text = text.replace("\xa0", " ").strip()  # x√≥a non-breaking space

    # N·∫øu c√≥ c·∫£ d·∫•u ',' v√† '.' ‚Üí b·ªè d·∫•u '.' (ngƒÉn c√°ch ngh√¨n), gi·ªØ ',' (th·∫≠p ph√¢n)
    if "," in text and "." in text:
        text = text.replace(".", "")
    text = text.replace(",", ".")  # th·ªëng nh·∫•t d·∫•u ph·∫©y ‚Üí ch·∫•m

    # T√¨m s·ªë th·ª±c
    match = re.search(r"\d+(?:\.\d+)?", text)
    if not match:
        return np.nan
    num = float(match.group())

    # Quy ƒë·ªïi v·ªÅ t·ª∑ ƒë·ªìng
    if "t·ª∑" in text:
        return num
    elif "tri·ªáu" in text:
        return num / 1000
    elif "ƒë" in text:
        return num / 1000000000
    else:
        # Kh√¥ng c√≥ ƒë∆°n v·ªã r√µ ‚Üí gi·∫£ ƒë·ªãnh l√† t·ª∑
        return num

# √Åp d·ª•ng
df["gi√° ti·ªÅn"] = df["gi√° ti·ªÅn"].apply(convert_price_to_billion)
df["gi√° ti·ªÅn"] = pd.to_numeric(df["gi√° ti·ªÅn"], errors="coerce")
df = df.rename(columns={"gi√° ti·ªÅn": "gi√° ti·ªÅn (t·ª∑)"})

df
```

```python
# c·ªôt di·ªán t√≠ch
def convert_area(value):
    if pd.isna(value):
        return np.nan
    text = str(value).lower().replace(",", ".").replace(" ", "")
    num = re.findall(r"[\d\.]+", text)
    if not num:
        return np.nan
    area = float(num[0])
    return area
df["di·ªán t√≠ch ƒë·∫•t"] = df["di·ªán t√≠ch ƒë·∫•t"].apply(convert_area)
df = df.rename(columns={"di·ªán t√≠ch ƒë·∫•t": "di·ªán t√≠ch ƒë·∫•t (m2)"})
df
```

```python
def clean_so_phong(x):
    if pd.isna(x) or str(x).strip() == "":
        return np.nan
    
    text = str(x).lower().strip()
    
    # N·∫øu c√≥ c·ª•m "nhi·ªÅu h∆°n"
    match = re.search(r"\d+(?:[.,]\d+)?", text)
    if "nhi·ªÅu h∆°n" in text or "tr√™n" in text:
        match = re.search(r"\d+(?:[.,]\d+)?", text)
        if match:
            num = float(match.group().replace(",", "."))
            return num + 1   # c·ªông th√™m 1 so v·ªõi s·ªë t√¨m ƒë∆∞·ª£c
        else:
            return np.nan
    
    # T√¨m s·ªë trong chu·ªói
    match = re.search(r"\d+(?:[.,]\d+)?", text)
    if match:
        num = match.group().replace(",", ".")
        try:
            return float(num)
        except:
            return np.nan
    
    return np.nan

# L√†m s·∫°ch
df["s·ªë ph√≤ng ng·ªß"] = df["s·ªë ph√≤ng ng·ªß"].apply(clean_so_phong)
df["s·ªë ph√≤ng v·ªá sinh"] = df["s·ªë ph√≤ng v·ªá sinh"].apply(clean_so_phong)
df
```

# X·ª≠ l√Ω Nan + ngo·∫°i lai

```python
def fillna_with_median(df, column):
    median_value = df[column].median()
    n_missing = df[column].isna().sum()

    print(f"\nüßπ C·ªôt: '{column}'")
    print(f" - S·ªë l∆∞·ª£ng NaN: {n_missing}")
    print(f" - Gi√° tr·ªã median d√πng ƒë·ªÉ thay th·∫ø: {median_value:.3f}")

    # Thay NaN = median
    df[column] = df[column].fillna(median_value)

    print(f" - S·ªë l∆∞·ª£ng NaN sau khi x·ª≠ l√Ω: {df[column].isna().sum()}")
    print("-" * 50)
    return df
```

```python
def handle_outliers_iqr(df, column):
    # T√≠nh c√°c gi√° tr·ªã th·ªëng k√™ c∆° b·∫£n
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    # T√¨m c√°c gi√° tr·ªã ngo·∫°i lai
    outliers = df[(df[column] < lower) | (df[column] > upper)]
    n_outliers = len(outliers)

    # In th√¥ng tin chi ti·∫øt
    print(f"\nüìä C·ªôt: '{column}'")
    print(f" - Q1 (25%): {Q1:.3f}")
    print(f" - Q3 (75%): {Q3:.3f}")
    print(f" - IQR: {IQR:.3f}")
    print(f" - Ng∆∞·ª°ng h·ª£p l√Ω: [{lower:.3f}, {upper:.3f}]")
    print(f" - Ngo·∫°i lai ph√°t hi·ªán: {n_outliers}")
    print(f" - Gi√° tr·ªã min tr∆∞·ªõc khi c·∫Øt: {df[column].min():.3f}")
    print(f" - Gi√° tr·ªã max tr∆∞·ªõc khi c·∫Øt: {df[column].max():.3f}")

    # C·∫Øt ng∆∞·ª°ng (Winsorization)
    df[column] = df[column].clip(lower=lower, upper=upper)

    # In l·∫°i min‚Äìmax sau khi x·ª≠ l√Ω
    print(f" - Gi√° tr·ªã min sau khi c·∫Øt: {df[column].min():.3f}")
    print(f" - Gi√° tr·ªã max sau khi c·∫Øt: {df[column].max():.3f}")
    print("-" * 60)

    return df
```
