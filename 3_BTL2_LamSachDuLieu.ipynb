```python
# cột giá tiền
import re
import numpy as np
import pandas as pd

def convert_price_to_billion(value):
    if pd.isna(value):
        return np.nan

    # Làm sạch ký tự lạ
    text = str(value).lower().strip()
    text = re.sub(r"[^0-9a-zA-ZÀ-ỹ,.\s]", "", text)
    text = text.replace("\xa0", " ").strip()  # xóa non-breaking space

    # Nếu có cả dấu ',' và '.' → bỏ dấu '.' (ngăn cách nghìn), giữ ',' (thập phân)
    if "," in text and "." in text:
        text = text.replace(".", "")
    text = text.replace(",", ".")  # thống nhất dấu phẩy → chấm

    # Tìm số thực
    match = re.search(r"\d+(?:\.\d+)?", text)
    if not match:
        return np.nan
    num = float(match.group())

    # Quy đổi về tỷ đồng
    if "tỷ" in text:
        return num
    elif "triệu" in text:
        return num / 1000
    elif "đ" in text:
        return num / 1000000000
    else:
        # Không có đơn vị rõ → giả định là tỷ
        return num

# Áp dụng
df["giá tiền"] = df["giá tiền"].apply(convert_price_to_billion)
df["giá tiền"] = pd.to_numeric(df["giá tiền"], errors="coerce")
df = df.rename(columns={"giá tiền": "giá tiền (tỷ)"})

df
```

```python
# cột diện tích
def convert_area(value):
    if pd.isna(value):
        return np.nan
    text = str(value).lower().replace(",", ".").replace(" ", "")
    num = re.findall(r"[\d\.]+", text)
    if not num:
        return np.nan
    area = float(num[0])
    return area
df["diện tích đất"] = df["diện tích đất"].apply(convert_area)
df = df.rename(columns={"diện tích đất": "diện tích đất (m2)"})
df
```

```python
def clean_so_phong(x):
    if pd.isna(x) or str(x).strip() == "":
        return np.nan
    
    text = str(x).lower().strip()
    
    # Nếu có cụm "nhiều hơn"
    match = re.search(r"\d+(?:[.,]\d+)?", text)
    if "nhiều hơn" in text or "trên" in text:
        match = re.search(r"\d+(?:[.,]\d+)?", text)
        if match:
            num = float(match.group().replace(",", "."))
            return num + 1   # cộng thêm 1 so với số tìm được
        else:
            return np.nan
    
    # Tìm số trong chuỗi
    match = re.search(r"\d+(?:[.,]\d+)?", text)
    if match:
        num = match.group().replace(",", ".")
        try:
            return float(num)
        except:
            return np.nan
    
    return np.nan

# Làm sạch
df["số phòng ngủ"] = df["số phòng ngủ"].apply(clean_so_phong)
df["số phòng vệ sinh"] = df["số phòng vệ sinh"].apply(clean_so_phong)
df
```

# Xử lý Nan + ngoại lai

```python
def fillna_with_median(df, column):
    median_value = df[column].median()
    n_missing = df[column].isna().sum()

    print(f"\n🧹 Cột: '{column}'")
    print(f" - Số lượng NaN: {n_missing}")
    print(f" - Giá trị median dùng để thay thế: {median_value:.3f}")

    # Thay NaN = median
    df[column] = df[column].fillna(median_value)

    print(f" - Số lượng NaN sau khi xử lý: {df[column].isna().sum()}")
    print("-" * 50)
    return df
```

```python
def handle_outliers_iqr(df, column):
    # Tính các giá trị thống kê cơ bản
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    # Tìm các giá trị ngoại lai
    outliers = df[(df[column] < lower) | (df[column] > upper)]
    n_outliers = len(outliers)

    # In thông tin chi tiết
    print(f"\n📊 Cột: '{column}'")
    print(f" - Q1 (25%): {Q1:.3f}")
    print(f" - Q3 (75%): {Q3:.3f}")
    print(f" - IQR: {IQR:.3f}")
    print(f" - Ngưỡng hợp lý: [{lower:.3f}, {upper:.3f}]")
    print(f" - Ngoại lai phát hiện: {n_outliers}")
    print(f" - Giá trị min trước khi cắt: {df[column].min():.3f}")
    print(f" - Giá trị max trước khi cắt: {df[column].max():.3f}")

    # Cắt ngưỡng (Winsorization)
    df[column] = df[column].clip(lower=lower, upper=upper)

    # In lại min–max sau khi xử lý
    print(f" - Giá trị min sau khi cắt: {df[column].min():.3f}")
    print(f" - Giá trị max sau khi cắt: {df[column].max():.3f}")
    print("-" * 60)

    return df
```
