```python
def lay_link_sp(product_url):
    current_page = 1 
    driver.get(product_url)
    sleep(random.randint(3,6))
    while True:
        try:
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            time.sleep(3)
    
            divs = WebDriverWait(driver, 10).until(
            EC.presence_of_all_elements_located((By.CSS_SELECTOR, "div.c15fd2pn"))
            )
    
            for div in divs:
                try:
                    a_tag = div.find_element(By.TAG_NAME, "a")
                    href = a_tag.get_attribute("href")
                    links.append(href)
                except:
                    continue
            print(f"Tìm được {len(links)} link")
            time.sleep(5)
            
            buttons = WebDriverWait(driver, 10).until(
                EC.presence_of_all_elements_located((By.CSS_SELECTOR, "button.Paging_redirectPageBtn__KvsqJ"))
            )
            print(f"Tìm thấy {len(buttons)} nút chuyển trang")
    
            next_btn = None
            for idx, btn in enumerate(buttons, start=1):
                print(f"Nút {idx}: outerHTML = {btn.get_attribute('outerHTML')}")
                try:
                    btn.find_element(By.CSS_SELECTOR, "i.Paging_rightIcon__3p8MS")
                    print("--> Đây là nút NEXT")
                    next_btn = btn
                    break
                except:
                    print("--> Không có icon right trong nút này")
                    continue
    
            if not next_btn or not next_btn.is_enabled():
                print("Không tìm thấy nút NEXT hoặc nút bị disable. Thoát vòng lặp.")
                break
    
            # click nút next
            driver.execute_script("arguments[0].click();", next_btn)
            current_page += 1
            print(f"Đang ở trang {current_page}")
    
            # đợi nội dung trang load
            sleep(random.randint(3,6))
    
        except (TimeoutException, ElementClickInterceptedException, NoSuchElementException):
            print("Không tìm thấy bất kỳ nút phân trang nào. Thoát vòng lặp.")
            break

    print(f"Tổng số trang: {current_page}")
    return current_page
```

```python
links = [] 
lay_link_sp("https://www.nhatot.com/mua-ban-bat-dong-san-ha-noi")
print(len(links))
```
